```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```python
training = r'C:\Users\jatin\Desktop\archive (1)\real_vs_fake\real-vs-fake\train' 
valid = r'C:\Users\jatin\Desktop\archive (1)\real_vs_fake\real-vs-fake\valid'
test = r'C:\Users\jatin\Desktop\archive (1)\real_vs_fake\real-vs-fake\test'
```

```python
image_gen = ImageDataGenerator(rescale=1./255.)
batch_size = 64
train_flow = image_gen.flow_from_directory(
    training,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='binary'
)
valid_flow = image_gen.flow_from_directory(
    valid,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='binary'
)
```

```python
cnn = tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[224, 224, 3]))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
cnn.summary()
```

```python
opt = tf.keras.optimizers.RMSprop(
        learning_rate=0.0001,
        momentum=0.01)
```

```python
cnn.compile(optimizer =opt , loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
train_steps = 40000//batch_size
valid_steps = 5000//batch_size
history = cnn.fit_generator(
    train_flow,
    epochs=20,
    steps_per_epoch = train_steps,
    validation_data = valid_flow,
    validation_steps = valid_steps
)
```

```python
image_gen = ImageDataGenerator(rescale=1./255.)
test_flow = image_gen.flow_from_directory(
    test,
    target_size=(224, 224),
    batch_size=1,
    shuffle=False,
    class_mode='binary'
)
```

```python
results = cnn.evaluate(test_flow)
```

```python
y_pred = cnn.predict(test_flow)
y_test = test_flow.classes
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
cm = confusion_matrix(y_test,y_pred.round())
plt.figure(dpi=300)
ax=plt.subplot()
sns.heatmap(cm,annot=True,fmt = "g",ax=ax)
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels');
ax.set_title('Confusion Matrix');
ax.xaxis.set_ticklabels(['Fake','Real']);ax.yaxis.set_ticklabels(['Fake','Real'])
```

```python
from sklearn import metrics
print("ROC AUC Score:", metrics.roc_auc_score(y_test, y_pred))
print("AP Score:", metrics.average_precision_score(y_test, y_pred))
print()
print(metrics.classification_report(y_test, y_pred > 0.5))
```

```python
total1=sum(sum(cm))
#####from confusion matrix calculate accuracy
accuracy1=(cm[0,0]+cm[1,1])/total1
print ('Accuracy : ', accuracy1)

sensitivity1 = cm[0,0]/(cm[0,0]+cm[0,1])
print('Sensitivity : ', sensitivity1 )

specificity1 = cm[1,1]/(cm[1,0]+cm[1,1])
print('Specificity : ', specificity1)
```

```python
def plot_accuracy(epochs, acc, val_acc):
    plt.figure(dpi=300)
    plt.plot(epochs, acc, 'bo', label='Training accuracy')
    plt.plot(epochs, val_acc, 'orange', label = 'Validation accuracy')
    plt.title('Training and Validation Accuracy')
    plt.legend()
    plt.show()
```
